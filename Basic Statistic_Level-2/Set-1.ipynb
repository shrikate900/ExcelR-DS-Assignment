```python
def outlier_detection(data,colname):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)
    iqr = q3-q1
    
    upper = q3+(1.5*iqr)
    lower = q1-(1.5*iqr)
    outliers=df['Measure X'][(df['Measure X']<=lower) |(df['Measure X']>=upper)]
    return outliers
```

```python
outlier_detection(df,'Measure X')
```
